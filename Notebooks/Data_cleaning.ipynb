```python
#Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
#drop useless and duplicates columns
df.drop(['id','imdb_id','homepage','budget','revenue','Unnamed: 21','tagline','keywords'],axis = 1,inplace = True)
```

```python
df.head()
```

```python
df.shape
```

```python
#Missing values by columns
df.isna().sum()
```

```python
#Drop rows with missing values
df.dropna(axis=0, inplace=True)
df.shape
```

```python
df.isna().sum()
```

```python
#keep the principal production company
df.production_companies = df.production_companies.str.split('|').str[0]
```

```python
df['cast'] = df.cast.str.split('|').apply(lambda x : ', '.join(x))
```

```python
#keep the 3 principal genres
df['genres'] = df.genres.str.split('|').str[0:3].apply(lambda x : ', '.join(x))
```

```python
df.head()
```

```python
#Unidecode
for col in ['original_title','cast','director','overview','genres']:
    df[col] = df[col].apply(lambda x : unidecode(x))
```

```python
#Convert objects columns to numeric
cols = ['popularity','runtime','vote_count']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
#Convert release_date to datetime
df['release_date'] = pd.to_datetime(df['release_date'],errors='coerce')
```

```python
#When the format is DD/MM/YY the year is not the good one ( ex : 01/01/66 is transformed in 01-01-2066)
def year_correction(date):
    current_date = datetime.datetime.now()
    if date > current_date:
        date = date.replace(year=date.year - 100)
    return date
```

```python
df['release_date'] = df['release_date'].apply(year_correction)
```

```python
#Missing values
df.loc[df.release_date.isna() == True]
```

```python
#Convert release_year to datetime and keep only good format dates
df['release_year'] = df['release_date'].dt.year.astype('O')
```

```python
#Creating columns for month and day released
df['release_day'] = df['release_date'].dt.day_name()
df['release_month'] = df['release_date'].dt.month_name()
```

```python
#drop missing values
df.dropna(axis=0, inplace=True)
df.shape
```

```python
df.dtypes
```

```python
#Creating a column profit
df['PROFIT'] = df['revenue_adj'] - df['budget_adj']
```

```python
#Renaming columns and reset index
df.rename(str.upper, axis=1, inplace=True)
df.rename(columns={"BUDGET_ADJ": "BUDGET", "REVENUE_ADJ": "REVENUE"}, inplace=True)
df.reset_index(drop = True, inplace=True)
```

```python
#reindexing columns
df = df.reindex(columns=[
 'ORIGINAL_TITLE',
 'CAST',
 'DIRECTOR',
 'OVERVIEW',
 'GENRES',
 'PRODUCTION_COMPANIES',
 'RUNTIME',
 'RELEASE_DATE',
 'RELEASE_YEAR',
 'RELEASE_MONTH',
 'RELEASE_DAY',
 'BUDGET',
 'REVENUE',
 'PROFIT',
 'VOTE_COUNT',
 'VOTE_AVERAGE',
 'POPULARITY'    
])
```
